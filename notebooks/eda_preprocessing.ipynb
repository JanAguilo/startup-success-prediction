```python
df = df.rename(columns=str.lower)
df = df.drop(columns=["unnamed: 0.1"])
```

```python
success = ["acquired", "ipo"]
failure = ["closed"]

df = df[df["status"].str.lower().isin(success + failure)]

df["success"] = df["status"].str.lower().map(
    {s:1 for s in success} | {f:0 for f in failure}
)
```

```python
date_cols = ["founded_at", "closed_at", "first_funding_at", "last_funding_at"]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")
```

```python
df["founded_year"] = df["founded_at"].dt.year
df["closed_year"] = df["closed_at"].dt.year

df["startup_age"] = df["closed_year"] - df["founded_year"]
df["startup_age"] = df["startup_age"].fillna(df["startup_age"].median())
```

```python
df["log_funding"] = np.log1p(df["funding_total_usd"])
```
